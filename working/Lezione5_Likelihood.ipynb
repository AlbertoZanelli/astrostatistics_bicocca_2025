```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Step 1: Generate fake data
np.random.seed(42)
true_mu = 5.0
known_sigma = 1.0
n_points = 100
data = np.random.normal(true_mu, known_sigma, n_points)

# Step 2: Define a grid of mu values for plotting
mu_grid = np.linspace(true_mu - 6*known_sigma, true_mu + 6*known_sigma, 500)

# Step 3: Compute individual likelihoods p(x_i | mu, sigma)
likelihoods = [norm.pdf(data_i, loc=mu_grid, scale=known_sigma) for data_i in data]

# Step 4: Compute joint likelihood (product of individual likelihoods)
joint_likelihood = np.prod(likelihoods, axis=0)

# Step 5: Plot individual likelihoods
plt.figure(figsize=(10, 6))
for i, likelihood in enumerate(likelihoods):
    plt.plot(mu_grid, likelihood)

# Plot the joint likelihood
plt.plot(mu_grid, joint_likelihood / np.max(joint_likelihood))

plt.title('Individual Likelihoods and Joint Likelihood')
plt.xlabel(r'$\mu$')
plt.ylabel('Likelihood')
plt.legend()
plt.grid(True)
plt.show()
```
